```python
import pandas as pd;
df = pd.read_csv('D:\Gamadata2 (no ssd)\Gamadata-Deira\gamadata-2\Full Data Garis Kemiskinan 2004 - 2024 with External Features.csv')
```

```python
# create a new dataframe for the mean values
mean_df = pd.DataFrame()
# iterating from 2004 to 2024
for i in range(2004, 2025):
    # calculate the mean of the 'Garis Kemiskinan' column for the current year
    mean_value = df[df['Tahun'] == i]['Garis Kemiskinan'].mean()
    # add the mean value to the new dataframe
    mean_df[f'{i}'] = [mean_value]
```

```python
mean_df
```

```python
# change the structure of the dataframe to have years as rows
mean_df = mean_df.T
# reset the index to have a clean dataframe
mean_df.reset_index(inplace=True)
mean_df.columns = ['Tahun', 'Garis Kemiskinan']
```

```python
import pandas as pd

# Assume mean_df and national_forecast are already defined

# Rename columns in mean_df to match national_forecast
mean_df_renamed = mean_df.rename(columns={
    "Tahun": "Year",
    "Garis Kemiskinan": "National Average Forecast"
})

# Concatenate both DataFrames
merged_df = pd.concat([mean_df_renamed, national_forecast], ignore_index=True)

# Sort by Year just in case (optional, but useful)
merged_df = merged_df.sort_values(by="Year").reset_index(drop=True)

# Show result
print(merged_df)
```

```python
# Save the merged DataFrame to a CSV file
merged_df.to_csv('merged_forecasts.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Split the dataframe
historical_df = merged_df[merged_df['Year'] < 2025]
forecast_df = merged_df[merged_df['Year'] >= 2025]

# Create the plot
plt.figure(figsize=(12, 6))

# Plot historical data
plt.plot(historical_df['Year'], historical_df['National Average Forecast'], label='Historical', color='blue', marker='o')
for x, y in zip(historical_df['Year'], historical_df['National Average Forecast']):
    plt.text(x, y, f'{y:,.0f}', fontsize=8, ha='center', va='bottom', color='blue')

# Plot forecast data
plt.plot(forecast_df['Year'], forecast_df['National Average Forecast'], label='Forecast', color='orange', marker='o')
for x, y in zip(forecast_df['Year'], forecast_df['National Average Forecast']):
    plt.text(x, y, f'{y:,.0f}', fontsize=8, ha='center', va='bottom', color='orange')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('National Average Poverty Line')
plt.title('National Average Forecast Over Time (2004–2029)')
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Apply a Seaborn style
sns.set(style="whitegrid")

# Create figure
plt.figure(figsize=(14, 7))

# Historical line
plt.plot(historical_df['Year'], historical_df['National Average Forecast'], 
         label='Historical', color='#1f77b4', linewidth=2.5, marker='o')
plt.fill_between(historical_df['Year'], historical_df['National Average Forecast'], 
                 color='#1f77b4', alpha=0.1)

# Forecast line (dashed)
plt.plot(forecast_df['Year'], forecast_df['National Average Forecast'], 
         label='Forecast', color='#ff7f0e', linewidth=2.5, linestyle='--', marker='o')
plt.fill_between(forecast_df['Year'], forecast_df['National Average Forecast'], 
                 color='#ff7f0e', alpha=0.1)

# Annotate values with 'Rp' and in 'K' (thousand)
def annotate(df, color):
    for x, y in zip(df['Year'], df['National Average Forecast']):
        y_rounded = round(y / 1000)  # Convert to thousands
        plt.text(x, y + 10000, f'Rp{y_rounded:,}K', fontsize=9, ha='center', color=color, fontweight='bold')

annotate(historical_df, '#1f77b4')
annotate(forecast_df, '#ff7f0e')

# Titles and labels
plt.title('📈 National Average Forecast (2004–2029)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('National Average Poverty Line', fontsize=12)
plt.xticks(historical_df['Year'].tolist() + forecast_df['Year'].tolist(), rotation=45)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Style improvements
plt.tight_layout()
plt.show()
```
